```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Đường dẫn tới thư mục data đã tổ chức xong ở bước trước
data_dir = 'train_organized'

def visualize_samples(data_dir, num_samples=10):
    # Lấy danh sách tất cả các tên xe (tên thư mục con)
    all_classes = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]
    
    plt.figure(figsize=(20, 10))
    
    for i in range(num_samples):
        # Chọn ngẫu nhiên một lớp xe
        random_class = random.choice(all_classes)
        class_path = os.path.join(data_dir, random_class)
        
        # Chọn ngẫu nhiên một tấm ảnh trong lớp đó
        random_image_name = random.choice(os.listdir(class_path))
        image_path = os.path.join(class_path, random_image_name)
        
        # Mở ảnh
        img = Image.open(image_path)
        
        # Hiển thị
        plt.subplot(2, 5, i + 1)
        plt.imshow(img)
        plt.title(random_class.replace('_', ' '), fontsize=10)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

visualize_samples(data_dir)
```
